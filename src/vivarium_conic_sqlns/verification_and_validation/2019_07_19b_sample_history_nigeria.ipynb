```python
history_data_path = '/share/costeffectiveness/results/sqlns/presentation/sample_history/2019_07_18_14_40_53/sample_history.hdf'
history_data = pd.read_hdf(history_data_path)
```

```python
cause_names=['lower_resipratory_infections', 'diarrheal_diseases', 'measles', 'iron_deficiency', 'protein_energy_malnutrition']

risk_names = ['iron_deficiency', 'child_stunting', 'child_wasting']
```

```python
@interact(simulant_id=Text(
              value='5643',
              placeholder='Type simulant id'),
          risk=Dropdown(
              options=risk_names,
              value='iron_deficiency'))

def plot_exposure(simulant_id, risk):
    try:
        simulant = history_data.loc[int(simulant_id)]
    except:
        raise ValueError(f'Simulant {simulant_id} not exist')
        
    sex = simulant.sex[0]
    age = round(simulant.age[0]*12, 1)
    start = simulant.sqlns_treatment_start.unique()
    end = simulant.sqlns_treatment_end.unique()
    xx = simulant.index
    
    plt.figure(figsize=(16,8))
    
    if risk == 'iron_deficiency':
        plt.plot(xx, simulant.iron_deficiency_exposure, label='treated', linewidth=2, drawstyle='steps-mid')
        plt.plot(xx, simulant.iron_deficiency_exposure_baseline, '--', label='baseline', linewidth=3, drawstyle='steps-mid')
        plt.ylabel('Hemoglobin\n$(g/L)$', fontsize=20)
    elif risk == 'child_stunting':
        plt.plot(xx, simulant.child_stunting_exposure-10, label='treated', linewidth=2, drawstyle='steps-mid')
        plt.plot(xx, simulant.child_stunting_exposure_baseline-10, '--', label='baseline', linewidth=3, drawstyle='steps-mid')
        plt.ylabel('HAZ', fontsize=20)
    else:
        plt.plot(xx, simulant.child_wasting_exposure-10, label='treated', linewidth=2, drawstyle='steps-mid')
        plt.plot(xx, simulant.child_wasting_exposure_baseline-10, '--', label='baseline', linewidth=3, drawstyle='steps-mid')
        plt.ylabel('WHZ', fontsize=20)

    if len(start) != 1:
        plt.axvline(start[1], color='k', linestyle='dashed')
        plt.axvline(end[1], color='k', linestyle='dashed')    
    plt.axvspan(pd.Timestamp('2020-01-01'), pd.Timestamp('2020-01-15'), alpha=0.1, color='blue', label='treatment unavailable')
    
    plt.title(f'{risk} exposure for simulant {simulant_id}, {sex}, {age} months old', fontsize=20)
    plt.xlabel('Time', fontsize=20)
    plt.xticks(rotation=45, fontsize=16)
    plt.yticks(fontsize=16)
    plt.legend(loc=(1.01, .01), fontsize=16)
    plt.grid()
```

```python
@interact(simulant_id=Text(
              value='5643',
              placeholder='Type simulant id'))

def plot_incidence(simulant_id):
    try:
        simulant = history_data.loc[int(simulant_id)]
    except:
        raise ValueError(f'Simulant {simulant_id} not exist')
    
    sex = simulant.sex[0]
    age = round(simulant.age[0]*12, 1)
    start = simulant.sqlns_treatment_start.unique()
    end = simulant.sqlns_treatment_end.unique()
    xx = simulant.index
    
    plt.figure(figsize=(16,8))
    
    for cause in cause_names[:3]:
        plt.plot(xx, getattr(simulant, f'{cause}_incidence_rate'), linewidth=2, label=cause)
        
    if len(start) != 1:
        plt.axvline(start[1], color='k', linestyle='dashed')
        plt.axvline(end[1], color='k', linestyle='dashed')    
    plt.axvspan(pd.Timestamp('2020-01-01'), pd.Timestamp('2020-01-15'), alpha=0.1, color='blue', label='treatment unavailable')
    
    plt.title(f'simulant {simulant_id}, {sex}, {age} months old', fontsize=20)
    plt.xlabel('Time', fontsize=20)
    plt.ylabel('Incidence', fontsize=20)
    plt.xticks(rotation=45, fontsize=16)
    plt.yticks(fontsize=16)
    plt.legend(loc=(1.01, .01), fontsize=16)
    plt.grid()
```

```python
def plot_multiple_simulants_exposure(N, risk):
    
    plt.figure(figsize=(16,8))
    
    for simulant_id in np.random.choice(history_data.index.levels[0], size=N, replace=False):
        simulant = history_data.loc[simulant_id]
        sex = simulant.sex[0]
        age = round(simulant.age[0]*12, 1)
        xx = simulant.index
        
        if risk == 'iron_deficiency':
            plt.plot(xx, simulant.iron_deficiency_exposure, drawstyle='steps-mid',
                     label=f'simulant {simulant_id}, {sex}, {age} months old')
            plt.ylabel('Hemoglobin\n$(g/L)$', fontsize=20)
        elif risk == 'child_stunting':
            plt.plot(xx, simulant.child_stunting_exposure-10, drawstyle='steps-mid',
                     label=f'simulant {simulant_id}, {sex}, {age} months old')
            plt.ylabel('HAZ', fontsize=20)
        else:
            plt.plot(xx, simulant.child_wasting_exposure-10, drawstyle='steps-mid',
                     label=f'simulant {simulant_id}, {sex}, {age} months old')
            plt.ylabel('WHZ', fontsize=20)

    plt.axvspan(pd.Timestamp('2020-01-01'), pd.Timestamp('2020-01-15'), alpha=0.1, color='blue', label='treatment unavailable')
    plt.title(f'{risk} exposure for {N} random simulants in treatment scenario', fontsize=20)
    plt.xlabel('Time', fontsize=20)
    plt.xticks(rotation=45, fontsize=16)
    plt.yticks(fontsize=16)
    plt.legend(loc=(1.01, .01), fontsize=16)
    plt.grid()
```

```python
plot_multiple_simulants_exposure(5, 'iron_deficiency')
```
